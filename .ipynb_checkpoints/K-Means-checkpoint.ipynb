```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,Y=make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

## Step-1 Initialisation

```python
k=5
color=["green","red","orange","blue","yellow"]

clusters={}

for i in range(k):
    x=np.random.randint(-12,12)
    y=np.random.randint(-7.5,10)
    
    center=np.array([x,y])
    points=[]
    
    cluster={
        "center":center,
        "color":color[i],
        "points":points
    }
    
    clusters[i]=cluster
```

```python
clusters
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

## Step-2 Assigning Points to nearest cluster

```python
def AssignPointsToCluster(clusters,X):
    for i in range(X.shape[0]):
        dist=[]
        
        for j in range(len(clusters)):
            d=distance(clusters[j]["center"],X[i])
            dist.append(d)
            
        cluster=np.argmin(dist)
        clusters[cluster]["points"].append(X[i])
```

```python
AssignPointsToCluster(clusters,X)
```

```python
for i in range(len(clusters)):
    print(len(clusters[i]["points"]))
```

## Step-3 Update Cluster `Centroids`

```python
def updateClusterCenter(Clusters):
    for i in range(len(clusters)):
        pnts=np.array(clusters[i]["points"])
        
        if pnts.shape[0] >0:
            new_center=np.mean(pnts,axis=0)
            clusters[i]["center"]=new_center
            clusters[i]["points"]=[]
```

```python
def plotClusters(Clusters):
    for i in range(len(clusters)):
        pnts=np.array(clusters[i]["points"])
        
        if(pnts.shape[0]>0):
            plt.scatter(pnts[:,0],pnts[:,1],c=clusters[i]["color"])
        
        plt.scatter(clusters[i]["center"][0],clusters[i]["center"][1],marker="*",c=clusters[i]["color"])
    plt.show()
```

```python
updateClusterCenter(clusters)
AssignPointsToCluster(clusters,X)
plotClusters(clusters)
```

## K-Means++
* k-Means is quite sensitive to initialization ,if the init is not good,our algorithm is not able to make desired number of clusters like we can see above 
* To overcome this problem we use technique called k-Means++ which chooses initial centers so that they are statically close to final ones
* The method is default choice of sklearn ,so we will use it direcly without going into much details

### Algorithm
<html>
    <ol>
        <li>Randomly select the first centroid from the data points.</li>
        <li>For each data point compute its distance from the nearest, previously choosen centroid.</li>
        <li>Select the next centroid from the data points such that the probability of choosing a point as centroid is directly proportional to its distance from the nearest, previously chosen centroid. (i.e. the point having maximum distance from the nearest centroid is most likely to be selected next as a centroid)</li>
        <li>Repeat steps 2 and 3 untill k centroids have been sampled</li>
    </ol>
</html>

# step-1 Initialisation rest steps are same

```python
k=5
color=["green","red","orange","blue","yellow"]

centroids=[]
centroids.append(X[np.random.randint(0,X.shape[0])])

for i in range(k-1):
    dist=[]
    
    for j in range(X.shape[0]):
        distances=((np.array(centroids)-X[j])**2).sum(axis=1)
        dist.append(distances.min())

    idx=(np.array(dist)).argmax()
    centroids.append(X[idx])
    

clusters={}
for i in range(k):
    center=centroids[i]
    points=[]
    
    cluster={
        "center":center,
        "color":color[i],
        "points":points
    }
    
    clusters[i]=cluster
```

```python
updateClusterCenter(clusters)
AssignPointsToCluster(clusters,X)
plotClusters(clusters)
```

```python
for i in range(k):
    print(clusters[i]['center'])
```
