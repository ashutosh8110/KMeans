# Dominant Color Extraction for Image Segmentation
* Image Segmentation
* Segmentation partitions an image into regions having similar visual appearance corresponding to parts of objects
* We will try to extract the most dominant 'K' Colors using K-Means
* We can apply K-Means with each pixel will reassigned to the closest of the K Colors, leading to segmentation


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=plt.imread("cat.png")
```

```python
X.shape
```

```python
X=X*255
X=X.astype('uint8')
plt.imshow(X)
plt.axis("off")
plt.show()
```

```python
all_pixels=X.reshape((-1,3))
```

```python
all_pixels.shape
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=5,)
```

```python
km.fit(all_pixels)
```

```python
centers=km.cluster_centers_.astype("uint8")
```

# Plot Dominant Colors

```python
centers
```

```python
for i in range(5):
    plt.subplot(3,5,i+1)
    a=np.zeros((100,100,3),dtype='uint8')
    a[:,:]=centers[i]
    plt.axis('off')
    plt.imshow(a)
```

# Segmenting Original Image

```python
new_img = np.zeros(X.shape, dtype="uint8").reshape((-1, 3))
```

```python
km.labels_
```

```python
for ix in range(new_img.shape[0]):
    new_img[ix] = centers[km.labels_[ix]]
```

```python
new_img=new_img.reshape(X.shape)
```

```python
plt.imshow(new_img)
plt.axis('off')
plt.show()
plt.imshow(X)
plt.axis('off')
plt.show()
```
